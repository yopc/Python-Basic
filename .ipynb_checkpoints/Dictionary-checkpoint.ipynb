Dictionary

A dictionary is a built-in data structure that stores data in key–value pairs.
It is mutable, dynamic, and stores unique keys.

```python
python_dict = {}
py_dict = dict()
```

Only immutable types: str, int, float, tuple, bool. used as key

```python
Ordered — dictionaries maintain insertion order starting from Python 3.7+.
Before that, they were unordered.
```

Merge two dictionary

```python
d1 = {"name": "John", "age": 25}
d2 = {"name": "kebede", "age": 35}
merged_dic = d1 | d2
print(merged_dic)
```
