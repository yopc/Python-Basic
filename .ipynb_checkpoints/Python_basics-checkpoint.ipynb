A set -> is an unordered collection of unique, immutable elements. Sets are:

* Mutable - you can add/remove elements
* Unordered - elements have no index/position
* Unique - no duplicate elements allowed
* Elements must be hashable (immutable types like strings, numbers, tuples)
set is important for
 -> membership testing
 -> removing duplicate



Creating empty set

```python
x = set() # x = {} will create dictionary not set 
x = {1,3,5,6}
print(x)
```

Duplicate will removed

```python
x = {1,3,5,5,6}
print(x)
```

```python
x.add(7)
x
```

```python
x.remove(3)   # error if not found
x.discard(3)  # no error if missing
x.pop()       # removes random item
x.clear()     # empties the set
x
```

SET OPERATION

```python
set_a = {1,4,2,6,7}
set_b = {8,3,9,10,4,6}
set_a | set_b   #  {1, 2, 3, 4, 6, 7, 8, 9, 10}
set_a & set_b   #  {4, 6}
set_a - set_b     #  {1, 2, 7}
```

```python
add_multiple = {4,5,6}
add_multiple.update([7,8,9])
add_multiple
```

.update() accepts lists, tuples, sets, strings, etc.
