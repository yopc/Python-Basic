A set -> is an unordered collection of unique, immutable elements. Sets are:

* Mutable - you can add/remove elements
* Unordered - elements have no index/position
* Unique - no duplicate elements allowed
* Elements must be hashable (immutable types like strings, numbers, tuples)
set is important for
 -> membership testing
 -> removing duplicate



Creating empty set

```python
x = set() # x = {} will create dictionary not set 
x = {1,3,5,6}
print(x)
```

Duplicate will removed

```python
x = {1,3,5,5,6}
print(x)
```

```python
x.add(7)
x
```

```python
x.remove(3)   # error if not found
x.discard(3)  # no error if missing
x.pop()       # removes random item
x.clear()     # empties the set
x
```

SET OPERATION

```python
set_a = {1,4,2,6,7}
set_b = {8,3,9,10,4,6}
set_a | set_b   #  {1, 2, 3, 4, 6, 7, 8, 9, 10}
set_a & set_b   #  {4, 6}
set_a - set_b   #  {1, 2, 7}  set difference set in once set but not in other 
set_a ^ set_b     # {1, 2, 3, 7, 8, 9, 10} symetric difference
```

Multiple set intersection 
set1 & set2 & set3
# OR
set1.intersection(set2, set3)


```python
add_multiple = {4,5,6}
add_multiple.update([7,8,9])
add_multiple
```

Pop do remove and return removed element from set

```python
set_a = {1,4,2,6,7}
set_a.pop()
```

```python
set_a = {1,4,2}
set_b = {1,4,2,5,6}
set_a.issubset(set_b)  # subset 
set_a <= set_b         # subset
set_a < set_b          # to check proper sub set
```

Removing multiple item from set 

```python
set_b = {1,4,2,5,6}
set_b -= {1,2,5}
set_b
```

```python
set_b = {1,4,2,5,6}
set_b.difference_update({1,2,5})
set_b
```

Iterating and removing directly causes RuntimeError. so how to bypass these issue 

```python
set_a = {1,4,2,5,6}
for x in set_a.copy():
    if x < 5:
        set_a.remove(x)
set_a
```
